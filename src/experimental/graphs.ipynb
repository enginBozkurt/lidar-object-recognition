```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.suptitle('RELu')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot([-1, 0, 1, 2], [0, 0, 1, 2])
```

```python
# Required Python Packages
def sigmoid(inputs):
 sigmoid_scores = [1 / float(1 + np.exp(- x)) for x in inputs]
 return sigmoid_scores
def line_graph(x, y, x_title, y_title):
 plt.plot(x, y)
 plt.xlabel(x_title)
 plt.ylabel(y_title)
 plt.grid(True)
 plt.suptitle("Sigmoid")
 plt.show()
graph_x = range(0,21)
graph_y = sigmoid(graph_x)
line_graph(graph_x, graph_y, "x", "y")
```
