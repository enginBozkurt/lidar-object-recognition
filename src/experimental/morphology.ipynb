```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('j.png')
img.shape
plt.imshow(img)
plt.show()
```

```python
kernel = (1, 1)
img_closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=3)
plt.imshow(img_closing)
plt.show()
```

```python
img = cv2.dilate(img_closing,kernel,iterations = 1)
plt.imshow(img_closing)
plt.show()
```
